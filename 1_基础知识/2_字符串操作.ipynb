```python
name = 'abcdefghijklmn'
# 字符串拼接 用加号操作
add_str = '123'+'345'
print(add_str)

# 占位符 %s是字符串，%d整型  %f浮点型,用%隔开，后面依次是变量
name = 'lisa'
age = 19
print('我的名字是%s，年龄是%d'%(name, age))

# 切片操作
str1 = name[0:4] #包头不包尾
print(str1)

#不写中开头位置或者终止位置，就是从头开始或者到尾部结束
str2 = name[2:]
str3 = name[:5]
print(str2)
print(str3)

# -1代表最后一个位置，-2代表倒数第二个位置，依次类推
str4 = name[0:-1]
print(str4)

# 第二个问号后。代表步长。即每次取字符的长度，步长是2，就是每两个字符取一下
str5 = name[0::2]
print(str5)

# 利用切片进行字符串反转,步长是-1。即从后往前取，每次取一个
str6 = name[9:0:-1]
print(str6)
```

```python
#字符串常见操作
my_str = "www.baidu.com www.taobao.com"
#
print(len(my_str))

# find是，这个字符串在原始字符串中的起始位置,返回第一个能匹配上的位置。即baidu的b这个字符在my_str中的位置。找不到返回-1
print(my_str.find("baidu"))
# rfind是从原始字符串右边开始查找
print(my_str.rfind("www"))
print(my_str.find("nostr"))

# startswith(xxx)检查字符串是否以xxx开头。能找到就返回True,找不到返回False
print(my_str.startswith("www"))

# endswith(xxx)检查字符串是否以xxx开头
print(my_str.endswith("com"))

# index、rindex和find、rfind大致一样，就是找不到 会报错，不是返回-1
print(my_str.index("baidu"))
print(my_str.rindex("www"))

# count 查找这个字符串在原始字符串中出现的次数
print(my_str.count('www'))

#replace 字符串替换.
my_str2 = my_str.replace("baidu", "jingdong") #将baidu替换为jingdong
print(my_str2)
my_str3 = my_str.replace("www", "qqq", 1) #替换一次，即替换第一个 
print(my_str3)

# split 字符串分割,以某个字符串作为分割
my_str4 = "ab,cd,ef,gh,ij,kl,mn"
print(my_str4.split(",")) # 以 , 作为分割，返回一个数组
print(my_str4.split(",", 3)) # 以 , 作为分割，最多分割成4份。

# capitalize 把字符串的第一个字符大写
print(my_str.capitalize())
# title 是把字符串的每个单词首字母大写
print(my_str.title())

#lower 把字符串中所有大写变为小写
print(my_str.lower())
# upper 把字符串中所有小写变为大写
print(my_str.upper())

# isalpha 判断字符串是否都是字母，返回bool类型
my_str6 = "abcd1"
print(my_str6.isalpha())
# isdigit 判断字符串是否都是数字，返回bool类型
print(my_str6.isdigit())
# isalnum 判断字符串是否仅包含数字或者字母，返回bool类型
print(my_str6.isalnum())
# isspace 判断字符串是否只包含空格，返回bool类型
print(my_str6.isspace())

# 把列表连接成字符串 split的逆应用
name1 = ["100", "200", "300"]

print("".join(name1))
print("-".join(name1)) # 用-连接
```
